```python

#pie chart with percentage

import matplotlib.pyplot as plt

food = ["Dosa","Idly","Poratta","Briyani"]
cost = [50,15,20,120]

plt.title("Food cost")
plt.pie(cost,labels=food,autopct="%0.1f%%", colors = ['Red','Green','Blue','Pink'])
plt.show()
```
