```python
#Handling duplicates

import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Alice'],
    'Age': [25, 30, 25],
    }

df = pd.DataFrame(data)

print(df.duplicated())
```

```python
print(df.drop_duplicates())
```

```python
#renaming columns

import pandas as pd
data = {
    "emp_nm":["john","mary","raj"],
    "emp_sal":[10000,20000,12000]
}
df = pd.DataFrame(data)

df.rename(columns={
    "emp_nm":"Employee_Name",
    "emp_sal":"Employee_Salary"
},inplace=True)

print(df)
```
