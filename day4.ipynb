```python

#median() --> middle value when sorted

import numpy as np

a = np.array([6,2,1,6,3,9])
median = np.median(a)
print(median)
```

```python
#Indexing & Slicing

import numpy as np
marks = np.array([90,80,75,60,100])
print(marks[1:4])
```

```python

#practice1

import numpy as np
scores = np.array([78,85,92,67,88])
total = np.sum(scores)
avg = np.mean(scores)
maximumscore = np.max(scores)
minimumscore = np.min(scores)

print(total)
print(avg)
print(maximumscore)
print(minimumscore)
```

```python

#practice2

sales = np.array([12000,15000,18000,14000,20000])
total_sales = np.sum(sales)
avg = np.mean(sales)
max = np.max(sales)
min = np.min(sales)

print(total_sales)
print(avg)
print(max)
print(min)
```

```python

#Reshape & Transpose

import numpy as np

a = np.array([1,2,3,4,5,6])
b = a.reshape(2,3)

print(b)

c= b.transpose()
print("\n",c)
```

```python



#Aggregation & Axis Operation

#sum() --> total

import numpy as np

a = np.array([1,2,3,4,5,6])
total = np.sum(a)
print(total)
```

```python


#mean --> average

import numpy as np

a = np.array([6,2,1,6,3,9])
avg = np.mean(a)
print(avg)
```
