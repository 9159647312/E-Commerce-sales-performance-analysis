```python
#practice: Attendence Calculator

def percentage(x,y):
  perc = (y/x)*100
  return round(perc,2 )

def status(percentage):
  if percentage >= 75:
    return "Eligible"
  else:
    return "Not eligible"

try:
  x = int(input("Total classes:"))
  y = int(input("classes attented:"))

  pct = percentage(x,y)
  st = status(pct)

  print("Percentage:",pct,"%")
  print("Status:",st)

except:
  print("Invalid input")

```
