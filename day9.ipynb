```python
import matplotlib.pyplot as plt


months = ["Jan","Feb","Mar","Apr"]
sales = [12000,18000,6000,22000]
plt.bar(months,sales,color=['Red','Green','Blue','Black'])
plt.xlabel("Months")
plt.ylabel("Sales")
plt.title("Sales Analysis")
plt.show()
```
