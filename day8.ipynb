```python
import matplotlib.pyplot as plt

import matplotlib.pyplot as plt


months = ["Jan","Feb","Mar","Apr"]
sales = [12000,18000,6000,22000]
plt.pie(sales,labels=months)
plt.xlabel("Months")
plt.ylabel("Sales")
plt.title("Sales Analysis")
plt.show()
```
